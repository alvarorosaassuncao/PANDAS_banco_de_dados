```python
# importando bibliotecas
import os
import pandas as pd
import plotly.express as px #biblioteca para criar gráficos


# Percorrer todos os arquivos da pasta de dados importando

lista_arquivo = os.listdir("/content/drive/MyDrive/Vendas")
display(lista_arquivo)


```

```python
#Criando uma tabela total para todas as planilhas 
tabela_total = pd.DataFrame()

# Importar as base de dados de vendas com for 

for arquivo in lista_arquivo: #Para cada arquivo na minha lista de arquivo
  if "Vendas" in arquivo: # se tem vendas no meu arquivo, então
    tabela = pd.read_csv(f"/content/drive/MyDrive/Vendas/{arquivo}")
    #complilando todas as tabelas para uma única tabela.
    tabela_total = tabela_total.append(tabela)
    display(tabela_total)

    
```

```python
#Passo 4 Calcular o produto mais vendido ( em quantidade)

tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[['Quantidade Vendida']].sort_values(by='Quantidade Vendida', ascending=False) #ascending=False ordena do maior para o menor
display(tabela_produtos)
```

```python
# Passo 5 Calcular o produto que mais faturou (em faturamento)
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']
display(tabela_total)
```

```python
tabela_faturamento = tabela_total.groupby('Produto').sum()
display(tabela_faturamento)
```

```python
#Passo 6 Calcular a loja/cidade que mais vendeu (em faturameto) - criar um grafico/ dashboard
tabela_lojas = tabela_total.groupby('Loja').sum() #Somando o faturamento por loja 
tabela_lojas = tabela_lojas[['Faturamento']] #selecionando apenas uma tabela 
display(tabela_lojas)
```

```python
import plotly.express as px

grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y="Faturamento")
grafico.show()
```
